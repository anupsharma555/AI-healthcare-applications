```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("clean_ccs_pr_icd10pcs.csv")

# Count the occurrences of each CCS_CATEGORY
category_counts = df['CCS_CATEGORY'].value_counts()

# Plot the counts as a bar chart
category_counts.plot(kind='bar', figsize=(10,5))
plt.xlabel('CCS_CATEGORY')
plt.ylabel('Count')
plt.title('Distribution of CCS_CATEGORY')
plt.tight_layout()
plt.show()
```

```python
# Extract the top ten CCS_CATEGORY values, their counts, and descriptions
top_10_categories = category_counts.head(10)
top_10_df = top_10_categories.reset_index()
top_10_df.columns = ['CCS_CATEGORY', 'count']
top_10_df = top_10_df.merge(
    df[['CCS_CATEGORY', 'CCS_CATEGORY_DESCRIPTION']].drop_duplicates(),
    on='CCS_CATEGORY',
    how='left'
)
print(top_10_df[['CCS_CATEGORY', 'count', 'CCS_CATEGORY_DESCRIPTION']])
```
